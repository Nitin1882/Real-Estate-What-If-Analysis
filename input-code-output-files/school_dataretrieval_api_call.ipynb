```python
#Import dependencies
import pandas as pd
from config import Api_key
import requests
import json
import time
```

```python
#API calls for pages 1-15.
base_url = "https://api.schooldigger.com/v1.2/rankings/schools/NJ?page="
appid = "522f0528"
tails_url = "&perPage=50&appID="
#url = f"{base_url}{page}{tails_url}{appid}&appKey={Api_key}"
#page1,2 data
#data = []
#page3,10 data
page1_15=[]
for page in range(1,16):
    page = page
    url = f"{base_url}{page}{tails_url}{appid}&appKey={Api_key}"
    print(f"fetching data for page:{page}")
    response = (requests.get(url)).json()
    #data.append(response)
    page1_15.append(response)
    time.sleep(60)
    
```

```python
#API calls for pages 16-28.
mineappid = "bf5e6ce7"
base_url = "https://api.schooldigger.com/v1.2/rankings/schools/NJ?page="
tails_url = "&perPage=50&appID="
page16_28=[]
for page in range(16,29):
    page = page
    url = f"{base_url}{page}{tails_url}{mineappid}&appKey={Api_key}"
    print(f"fetching data for page:{page}")
    response = (requests.get(url)).json()
    page16_28.append(response)
    time.sleep(60)

```

```python
print(json.dumps(response, indent =4))
```

```python
#Extracting data from pages 1 to 15 in dictionary

complete_list_schools = []

for i in range(len(page1_15)):
    school_list=page1_15[i]["schoolList"]
    for x in school_list:
        school = {}
        school["school_id"] = x["schoolid"]
        try:
            #school["latitude"] = x["address"]["latLong"]["latitude"]
            #school["longitude"] = x["address"]["latLong"]["longitude"]
            school["name"] = x["schoolName"] 
            school["rank_star"] = x["rankHistory"][0]["rankStars"]
            school["avg_std_score"] =x["rankHistory"][0]["averageStandardScore"]
            school["county"] = x["county"]["countyName"]
            school["zip"] = x["address"]["zip"]
        except KeyError:
            print(school["school_id"] + " is missing latitude.")
        
        
        
        
        complete_list_schools.append(school)
        

print(complete_list_schools)
```

```python
page16_28_df = pd.DataFrame(complete_list_schools2)
page16_28_df.head(30)
```
